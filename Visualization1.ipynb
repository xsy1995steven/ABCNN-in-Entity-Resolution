```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
length = [x for x in range(10,80,10)]
f1 = [95.7, 96.43, 96.82, 96.66, 97.01, 97.39, 97.55]
f2 = [96.79, 97.29, 97.15, 97.42, 97.88, 97.61, 98.03]
plt.figure(dpi=600)
plt.plot(length, f1,marker = 'o',label = 'Nave CNN')
plt.plot(length, f2, marker = 'x',label = 'ABCNN')
plt.title('F1 Scores of CNN Based Models with Truncated Dimensions')
plt.xlabel('Truncated Sentence Length')
plt.ylabel('F1 Score')
for a, b in zip(length, f1):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
for a, b in zip(length, f2):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
plt.legend()
plt.show()
```

```python
length = [x for x in range(10,80,10)]
run_time = [1.3, 1.6, 1.8, 2.3, 2.6, 2.9, 3.1]
run_time_2 = [96.5, 120.4, 155.6, 162.8, 189.9, 243.6, 286.5]
plt.figure(dpi=600)
plt.plot(length, run_time,marker = 'o',label = 'Nave CNN')
plt.plot(length, run_time_2, marker = 'x',label = 'ABCNN')
plt.yscale('log')
plt.title('Training Time of CNN Based Models with Truncated Dimensions')
plt.xlabel('Truncated Sentence Length')
plt.ylabel('Training Time')
plt.ylim(0.1, 500)
for a, b in zip(length, run_time):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
for a, b in zip(length, run_time_2):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
plt.legend()
plt.show()
```
